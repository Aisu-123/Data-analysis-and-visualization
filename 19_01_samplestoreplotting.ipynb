```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Sample - Superstore.csv",encoding="latin1")
df
```

```python
df1=df.drop(["Order ID","Customer ID","Customer Name","Row ID","Product ID"],axis=1)
```

```python
df1.head()
```

```python
df1.groupby(["City"]).max()["Profit"].sort_values(ascending=False)

```

lafayette has the highest profit city

```python
df1.groupby(["Region"]).max()["Profit"].sort_values(ascending=False)
```

central had the maximum profit region

```python
df1.groupby(["State"]).max()["Profit"].sort_values(ascending=False)
```

indiana has the maximum profit state

```python
df1.groupby(["Region"]).min()["Profit"].sort_values()
```

east  had the minimum profit region

```python
df1.groupby(["Category"]).count()["Sales"].sort_values(ascending=False)
```

office supplies was sold more

```python
df1.groupby(["Sub-Category"]).count()["Sales"].sort_values(ascending=False)
```

binders was sold more in sub-category


```python
df1.groupby(["Segment"]).sum()["Sales"].sort_values(ascending=False)
```

consumer segment brought more sales

```python
df1.groupby(["Segment"]).sum()["Profit"].sort_values(ascending=False)
```

consumer segment brought more profit
